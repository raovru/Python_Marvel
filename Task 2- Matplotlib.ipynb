```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA 
#Linear dimensionality reduction using Singular Value Decomposition of the data to project it to a lower dimensional space.
```

```python
%matplotlib inline 
#Jupyternotebook instead of plt.show() each time
```

```python
#declaring our x and y values
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Basic Plot
plt.plot(x, y, 'r-') #colour
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #Row. column, no. of plot
plt.plot(x, y, 'r')

#Intersection of x and y zeroes
plt.xlim(0, 5)
plt.ylim(0, 25)

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

```python
#Object Oriented method
fig = plt.figure() #It is a blank canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)

#Set axes label and limit - 1,2
axes.set_xlim(0, 5)
axes.set_ylim(0, 25)
axes.set_xlabel(' X Label')
axes.set_ylabel(' Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_ylim(0, 25)
axes1.set_xlim(0, 5)

axes2.set_xlim(0, 25)
axes2.set_ylim(0, 5)
```

```python
#Figure with multiple plots using subplot
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for cur_axes in axes:
    cur_axes.plot(x, y)
    
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(5, 5))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig('my_plot.png') #Save plot - 5
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x/3, label = 'X vs X/3')
ax.plot(x, x, label = 'X vs X itself')

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

#Added legend to the plot - 4
ax.legend(loc = 0) 
#Check documentation - 0 is for the best location
```

```python
iris_df = pd.read_csv('Iris.csv')
```

```python

plt.scatter(x = [iris_df.SepalLengthCm], y = [iris_df.PetalLengthCm], s=360, c='Red', alpha = 0.3)
plt.scatter(x = [iris_df.SepalWidthCm], y = [iris_df.PetalWidthCm], s =250, c = 'Indigo', alpha = 0.4)

plt.xlabel('Sepal Characteristics')
plt.ylabel('Petal Characteristics')
plt.title('Sepal vs Petal')

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(iris_df.SepalLengthCm, iris_df.SepalWidthCm, width = 0.085, color = 'cyan')

plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Width vs Sepal Length")
```
